```python
import pandas as pd
```

```python
df.fillna(df.mean(),inplace=False)
```

```python
df.fillna(df.mean(),inplace=False,subset=df.select_dtypes(include='float64'))
```

```python
floatvalue=df.select_dtypes(include='float64')
df.fillna(floatvalue.mean(),inplace=False)
```

```python
df1=df.select_dtypes(include='float64')
df.fillna(df1.median(), inplace=False)
```
